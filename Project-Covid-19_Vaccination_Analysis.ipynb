# importing pandas library for reading csv file in dataframe


```python
import pandas as pd
df=pd.read_csv('country_vaccinations.csv')
df
```

# dropping the rows which values are not available

```python
df.dropna(inplace=True)
```

# displaying dataframe

```python
df
```

# extracting year from date column

```python
df['year']= pd.to_datetime(df["date"], format="%Y-%m-%d").dt.year
```

# displaying year column

```python
df
```

# extracting month from date column

```python
df['month']= pd.to_datetime(df["date"], format="%Y-%m-%d").dt.month
df
```

# extracting day from date column

```python
df['day']= pd.to_datetime(df["date"], format="%Y-%m-%d").dt.day
df
```

# removing date column after extracting day,month and year

```python
df.drop(['date'],axis='columns',inplace=True)
df
```

# removing total_vaccination column

```python
inputs=df.drop('total_vaccinations',axis='columns')
inputs
```

# assigning total_vaccination column to target dataframe

```python
target=df['total_vaccinations']
```

# displaying the target

```python
target
```

# importing labelencoder library

```python
from sklearn.preprocessing import LabelEncoder
l_location=LabelEncoder()
l_vaccine=LabelEncoder()
```

# converting location and vaccine string values into integer 

```python
inputs['location_n']=l_location.fit_transform(inputs['location'])
inputs['vaccine_n']=l_vaccine.fit_transform(inputs['vaccine'])
```

# displaying location_n and vaccine_n column after encoding

```python
x=inputs.drop(['month','year','day'],axis='columns')
with pd.option_context('display.max_rows',None,'display.max_columns',None,'display.precision',5,):
    print(x)

               
```

# displaying inputs dataframe

```python
inputs
```

# removing location and vaccine column having string values

```python
new_inputs=inputs.drop(['location','vaccine'],axis='columns')
new_inputs
```

# importing required library for plotting graph

```python
import matplotlib.pyplot as plt
plt.xlabel('Location')
plt.ylabel('total_vaccination')
plt.scatter(new_inputs['location_n'],target, color="green", marker='+')
```

# splitting the data, 80% training data and 20% test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(new_inputs,target,test_size=0.2)
```

# importing library for linearregression

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model
```

# implementing linearregression model with training data

```python
model.fit(X_train,y_train)
```

# predicting the output with test data

```python
model.predict(X_test)
```

# checking the efficiency of linearregression model

```python
model.score(X_test,y_test)
```
